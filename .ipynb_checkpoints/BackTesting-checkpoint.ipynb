```python
import numpy as np
import pandas as pd
```

```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
import matplotlib.pyplot as plt
from bokeh.io import export_png

from backtesting.test import SMA, GOOG




class SmaCross(Strategy):
    def init(self):
        price = self.data.Close
        self.ma1 = self.I(SMA, price, 10)
        self.ma2 = self.I(SMA, price, 20)

    def next(self):
        if crossover(self.ma1, self.ma2):
            self.buy()
        elif crossover(self.ma2, self.ma1):
            self.sell()
```

```python
def myparser(dateTime):
    data = []
    for d in dateTime:
        
        data.append(d[:-12])
    return data
```

```python
file_name="AUD_CAD.csv"
data=pd.read_csv(file_name,index_col="Date",parse_dates=["Date"],date_parser=myparser)
data=data.loc[:,["Open","High","Low","Close","Volume"]].copy()
data
```

```python
bt = Backtest(data, SmaCross, commission=.002,
              exclusive_orders=True)
stats = bt.run()
a=bt.plot()
```
